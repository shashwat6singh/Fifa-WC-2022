```python
#Changing date datatype from object to date
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
#Changing datatypes from object to string
df = df.astype({"team1": "string", "team2": "string", "category": "string"})
df.info(verbose = True)
```

```python
#Changing possession datatype to int and removing % symbol
df['possession team1'] = df['possession team1'].str.replace('%','').astype(int)
df['possession team2'] = df['possession team2'].str.replace('%','').astype(int)
df.info(verbose = True)
```

```python
df
```

```python
df['possession in contest'] = df['possession in contest'].str.replace('%','').astype(int)
```

```python
df
```

```python
#Inspecting Missing Value in Dataset
df.isnull().sum().sort_values(ascending = False)
```

```python
#Removing Unwanted Columns
df.drop(columns = ['own goals team1','own goals team2','crosses completed team1','crosses completed team2','crosses team1','crosses team2'], inplace = True)
df.shape
```

```python
#Group A Analysis
GroupA = df.loc[df.category == 'Group A']
GroupA
```

```python
np.where(GroupA["number of goals team2"] == 0)
```

```python
#Distribution of Left & Right Channel Attacks for Teams in Group A
plt.figure(figsize = [20,10])

plt.subplot(2,2,1)
plt.hist(GroupA['left channel team1'], edgecolor ='green')
plt.xlabel('left channel team1')
plt.ylabel('density')

plt.subplot(2,2,2)
plt.hist(GroupA['left channel team2'], edgecolor ='green')
plt.xlabel('left channel team2')
plt.ylabel('density')

plt.subplot(2,2,3)
plt.hist(GroupA['right channel team1'], edgecolor ='green')
plt.xlabel('right channel team1')
plt.ylabel('density')

plt.subplot(2,2,4)
plt.hist(GroupA['right channel team1'], edgecolor ='green')
plt.xlabel('right channel team2')
plt.ylabel('density')

plt.show()
```

```python
#Checking the skewness to determine if the different channels in Group A has normal distirbution
left_team1 = GroupA['left channel team1'].skew()
left_team2 = GroupA['left channel team2'].skew()
right_team1 = GroupA['right channel team1'].skew()
right_team2 = GroupA['right channel team2'].skew()
left_team1, left_team2, right_team1, right_team2
```

```python
#Correlation between different measures of team 1
selected_columns = ['total attempts team1', 'number of goals team1', 'goal inside the penalty area team1', 'goal outside the penalty area team1', 'on target attempts team1', 'off target attempts team1', 'fouls against team1', 'offsides team1']
selected_df = df[selected_columns]
correlation_matrix = selected_df.corr()
print(correlation_matrix)
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(selected_columns)), selected_columns, rotation=90)
plt.yticks(range(len(selected_columns)), selected_columns)
plt.title('Correlation Matrix of Selected Columns')
plt.show()
```

```python
#Correlation between different measures of team 2
selected_columns = ['total attempts team2', 'number of goals team2', 'goal inside the penalty area team2', 'goal outside the penalty area team2', 'on target attempts team2', 'off target attempts team2', 'fouls against team2', 'offsides team2']
selected_df = df[selected_columns]
correlation_matrix = selected_df.corr()
print(correlation_matrix)
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(selected_columns)), selected_columns, rotation=90)
plt.yticks(range(len(selected_columns)), selected_columns)
plt.title('Correlation Matrix of Selected Columns')
plt.show()
```
